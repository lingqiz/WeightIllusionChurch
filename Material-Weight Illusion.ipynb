```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline  
# plt.style.use('ggplot')
```

```python
def posterior_prior(post_str):
    count_R0 = float(post_str.count('0'))
    count_R1 = float(post_str.count('1'))
    count_R2 = float(post_str.count('2'))
    total = count_R0 + count_R1 + count_R2
    
    per_R0 = count_R0 / total
    per_R1 = count_R1 / total
    per_R2 = count_R2 / total 
    
    N = 1
    P0 = (per_R0)

    width = 0.15       # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(0.1, P0, width, color = 'forestgreen')

    P1 = (per_R1)
    rects2 = ax.bar(0.6, P1, width, color = 'goldenrod')
    
    P2 = (per_R2)
    rects3 = ax.bar(1.1, P2, width, color = 'darkgray')

    # add some text for labels, title and axes ticks
    ax.set_ylabel('P(R)')
    ax.set_title('Posterior of Prior')

    ax.legend((rects1[0], rects2[0], rects3[0]), ('P0', 'P1', 'P2'))
    plt.show()
```
